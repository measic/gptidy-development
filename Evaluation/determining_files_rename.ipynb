```python
from item_type import ItemType 
NUM_FILES = 6098
SAMPLE_SIZE = 374
type = ItemType.ASSIGN
ignore_underscores = True
```

```python
# read in all files in the folder all_cells
all_cells = []
for i in range(NUM_FILES):
    with open(f'all_cells/{i}.py', 'r') as f:
        all_cells.append(f.read())
```

```python
import ast_determine_usable_items

desired_files = []

for i in range(NUM_FILES):
    try:
        usable_items = ast_determine_usable_items.determine_usable_items(all_cells[i], type, ignore_underscores)
    except SyntaxError:
        usable_items = []
    if len(usable_items) > 0:
        desired_files.append((i, usable_items))

print(desired_files)
```

```python
len(desired_files)
```

```python
# checking only: determine if any starts with an underscore
for i, item in enumerate(desired_files):
    for var in item[1]:
        if var.startswith("_"):
            print(item[0], var)
            break
```

```python
# flatten the tuple
desired_files_flattened = [(file_id, name) for file_id, names in desired_files for name in names]
desired_files_flattened
```

```python
len(desired_files_flattened)
```

```python
# select a random sample
import random
random.seed(42)
random.shuffle(desired_files_flattened)
random_sample = desired_files_flattened[:SAMPLE_SIZE]
```

```python
# save random selection of cells to a folder
import os
if not os.path.exists('random_samples'):
    os.makedirs('random_samples')

# save files
for i, file_id in enumerate(random_sample):
    with open(f'random_samples/{i}.py', 'w') as f:
        f.write(all_cells[file_id[0]])

# save data on file id
with open('random_samples/file_ids.txt', 'w') as f:
    for file_id in random_sample:
        f.write(f'{file_id[0]}\n')

# save data on name
with open('random_samples/names.txt', 'w') as f:
    for file_id in random_sample:
        f.write(f'{file_id[1]}\n')
```
