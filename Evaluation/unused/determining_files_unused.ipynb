```python
NUM_FILES = 6095
SAMPLE_SIZE_VARIABLES = 305
SAMPLE_SIZE_FUNCTIONS = 209
```

```python
# read in all files in the folder all_cells
all_cells = []
for i in range(NUM_FILES):
    with open(f'all_cells/{i}.py', 'r') as f:
        all_cells.append(f.read())
```

```python
# CELL IS SPECIFIC TO VULTURE
import subprocess

def get_file_ids_vulture(filter_func):
    desired_file_ids = []
    for i in range(NUM_FILES):
        file_name = f'all_cells/{i}.py'
        result = subprocess.run(['vulture', file_name], capture_output=True, text=True)
        outputs = result.stdout.strip().split('\n')
        outputs = filter_func(outputs)
        for output in outputs:
            file_number = int(output.split('/')[1].split(".")[0])
            if file_number not in desired_file_ids:
                desired_file_ids.append(file_number)
            break
    return desired_file_ids

var_filter = lambda outputs: [line for line in outputs if 'unused variable' in line]
func_filter = lambda outputs: [line for line in outputs if 'unused function' in line or 'unused method' in line]

desired_file_ids_variables = get_file_ids_vulture(var_filter)
desired_file_ids_functions = get_file_ids_vulture(func_filter)
```

```python
print(desired_file_ids_variables)
print(desired_file_ids_functions)
print(len(desired_file_ids_variables))
print(len(desired_file_ids_functions))
```

```python
import sys
sys.path.append('../')
from utils import num_tokens_from_string

# filter out desired files that are above token limits
top_limit = 1950

desired_file_ids_variables_filtered = [id for id in desired_file_ids_variables if num_tokens_from_string(all_cells[id]) < top_limit]
desired_file_ids_functions_filtered = [id for id in desired_file_ids_functions if num_tokens_from_string(all_cells[id]) < top_limit]

print(len(desired_file_ids_variables_filtered))
print(len(desired_file_ids_functions_filtered))
```

```python
# select a random sample
import random
random.seed(42)
random.shuffle(desired_file_ids_variables_filtered)
random.shuffle(desired_file_ids_functions_filtered)

random_sample_variables = desired_file_ids_variables_filtered[:SAMPLE_SIZE_VARIABLES]
random_sample_functions = desired_file_ids_functions_filtered[:SAMPLE_SIZE_FUNCTIONS]
```

```python
# save random selection of cells to a folder
def save_random_sample(file_name, files):
    import os
    if not os.path.exists(file_name):
        os.makedirs(file_name)

    for i, file_id in enumerate(files):
        with open(f'{file_name}/{i}.py', 'w') as f:
            f.write(all_cells[file_id])

save_random_sample('random_samples_variables', random_sample_variables)
save_random_sample('random_samples_functions', random_sample_functions)
```
