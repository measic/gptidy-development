```python
NUM_FILES = 6095
SAMPLE_SIZE = 350
```

```python
# read in all files in the folder all_cells
all_cells = []
for i in range(NUM_FILES):
    with open(f'all_cells/{i}.py', 'r') as f:
        all_cells.append(f.read())
```

```python
# determine which files have formatting issues
import sys
sys.path.append('../')
from common import IGNORE_TYPES
import subprocess

desired_file_ids = []

for i in range(NUM_FILES):
    file_name = f'all_cells/{i}.py'
    result = subprocess.run(['pycodestyle', f'--ignore={IGNORE_TYPES}', '--statistics', file_name], stdout=subprocess.PIPE).stdout.decode('utf-8')
    result = result.split('\n')
    # keep only the lines that start with an integer
    result = [line for line in result if line and line[0].isdigit()]
    # store the error counts in a hash
    if len(result) > 0 and i not in desired_file_ids:
        desired_file_ids.append(i)
```

```python
print(desired_file_ids)
print(len(desired_file_ids))
```

```python
import sys
sys.path.append('../../')
from utils import num_tokens_from_string

# filter out desired files that are above token limits
top_limit = 1950

desired_file_ids_filtered = [id for id in desired_file_ids if num_tokens_from_string(all_cells[id]) < top_limit]

print(len(desired_file_ids_filtered))
```

```python
# select a random sample
import random
random.seed(42)
random.shuffle(desired_file_ids_filtered)

random_sample = desired_file_ids_filtered[:SAMPLE_SIZE]
```

```python
# save random selection of cells to a folder
def save_random_sample(file_name, files):
    import os
    if not os.path.exists(file_name):
        os.makedirs(file_name)

    for i, file_id in enumerate(files):
        with open(f'{file_name}/{i}.py', 'w') as f:
            f.write(all_cells[file_id])

save_random_sample('random_samples_formatting', random_sample)
```
