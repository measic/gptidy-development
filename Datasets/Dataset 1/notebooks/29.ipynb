```python
import pandas as pd
import numpy as np
```

### Import the csv:

```python
movies = pd.read_csv("https://raw.githubusercontent.com/WilliamJCole/IS352_Week_7/master/movie_reviews.csv")
movies
```

### Average the ratings:

```python
movies = movies.set_index('Reviewers')
revavg = movies.mean(axis=1).round(decimals=0)

movieavg = movies.mean().round(decimals=0)

revavg
```

```python
movieavg
```

### Now we must normalize the ratings:

##### Here is a calculation that is used to calculate normalization:

![alt text](http://bi-analytics.org/uploads/monthly_2017_02/norm.png.734e9f7c5ed4fd68a5b5b9a74309af80.png)

```python
movies_tp = movies.T
moviesnormal = (movies_tp - movies_tp.min()) / (movies_tp.max() - movies_tp.min())
moviesnormal
```

#### Then we must split up the reviews by user and movies

```python
movies_norm = moviesnormal.mean(axis=1).round(decimals=5)
user_norm = moviesnormal.mean().round(decimals=5)

user_norm
```

```python
movies_norm
```

#### Now we must average the ratings:

```python
movies_norm = moviesnormal.mean(axis=1).round(decimals=5)
user_norm = moviesnormal.mean().round(decimals=5)

user_norm
```

## Extra Credit: Standardization:

#### Here is the calculation for Standardizaiton:

![alt text](http://bi-analytics.org/uploads/monthly_2017_02/stand.png.8afcd54950adc5134383de8bb625ebc4.png)


```python
moviesstd = (movies_tp - movies_tp.mean()) / (movies_tp.std())
moviesstd
```

```python
moviesavg_std = moviesstd.mean(axis=1).round(decimals=5)
useravg_std = moviesstd.mean().round(decimals=5)

useravg_std
```
